## Files and Exceptions

> Most of your programs conform to the *input-process-output model*:
> data comes in, gets manipulated, and then is stored, displayed, printed, or transferred

Using the `open()` BIF you can read files this is **line based**

```python
the_file = open('sketch.txt')
# Do something with the data in "the file"
the_file.close()
```

````python
import os  # import os from standard library

os.getcwd()  # ? get current directory
# os.makedirs("data") #? make directory
# os.chdir("data")  #? change to the that contains your data file

try:
    data = open("sketch.txt")

    data.seek(0)  # return to start of file
    tell = data.tell()
    print(tell)

    try:  # code block here is "protected"
        for line in data:
            if line.find(":") > 0:  # handle case where line has no colons
                (role, line_spoken) = line.split(":", 1)  #! see help for split below
                print("\x1b[1;32m" + role, end="")
                print("\x1b[0m said: \x1b[21;33m", end="")
                print(line_spoken, end="")
            else:
                print("\x1b[1;31m" + line)
    except ValueError:  # if runtime error occurs, this code block is executed
        pass  # ignore error with this python keyword

    data.close()  # close since we are done

except IOError:
    print("The data file is missing")
````

## Persistence of Data

````python
import os  # import os from standard library

os.getcwd()  # ? get current directory

try:
    data = open("sketch.txt")

    data.seek(0)  # return to start of file
    tell = data.tell()
    print(tell)

    # initialize empty lists
    man = []
    other = []

    try:  # code block here is "protected"
        for line in data:
            if line.find(":") > 0:  # handle case where line has no colons
                (role, line_spoken) = line.split(":", 1)  #! see help for split below
                # create empty lists for man and other
                line_spoken = line_spoken.strip()  # remove whitespace from line
                if role == "Man":
                    man.append(line_spoken)
                elif role == "Other Man":
                    other.append(line_spoken)

                print("\x1b[1;32m" + role, end="")
                print("\x1b[0m said: \x1b[21;33m", end="")
                print(line_spoken)
            else:
                print("\x1b[1;31m" + line)
    except ValueError:  # if runtime error occurs, this code block is executed
        pass  # ignore error with this python keyword

    data.close()  # close since we are done

except IOError:
    print("The data file is missing")

print("\n\x1b[1;36m")
print(man)
print("\x1b[1;35m")
print(other)
````

### The `finally` Keyword in Python

#### Overview
- The `finally` block in Python is used in conjunction with `try` and `except` blocks.
- It is intended to define clean-up actions that must be executed under all circumstances, regardless of whether an exception was raised or not.

#### Syntax
```python
try:
    # Code that might raise an exception
    risky_code()
except SomeException as e:
    # Code to handle the exception
    handle_exception(e)
finally:
    # Code that will always execute, regardless of what happened in try/except
    cleanup_actions()

````python
try:
    data = open("missing.txt")
    print(data.readline(), end="")
except IOError as err:
    print("File error " + err)
finally:
    if "data" in locals():
        data.close()
````

````python
""" AI version """

import os


def process_file(file_path):
    # Read, trim, split, and process data from a single file
    with open(file_path, "r") as file:
        data = file.readline().strip().split(",")
    return data


def process_all_files(directory):
    # Process all text files in the given directory
    processed_data = {}

    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            file_path = os.path.join(directory, filename)
            data = process_file(file_path)
            processed_data[filename] = data

    return processed_data


def sanitize(time_string):
    if "-" in time_string:
        splitter = "-"
    elif ":" in time_string:
        splitter = ":"
    else:
        return time_string

    (min, secs) = time_string.split(splitter)
    return min + "." + secs


def main():
    directory = "./coach_data"  # Replace with your directory path
    all_data = process_all_files(directory)

    for filename, data in all_data.items():
        # sanitized_data = []
        # for datum in data:
        #     sanitized_data.append(sanitize(datum))
        #! use list comprehension
        sanitized_data = set([sanitize(datum) for datum in data])
        sanitized_data = sorted(sanitized_data)[0:3]
        print(
            f"Data from {filename}: {sanitized_data}"
        )  # using list slices to get top 3 times


if __name__ == "__main__":
    main()
````
