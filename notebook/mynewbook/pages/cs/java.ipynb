#### Identifiers

Any name in Java is termed a **identifier**

Here are the rules:

- consists of letters digits, underscores `_` and dollar signs `$`
- can't start with digit
- can't be reserved keyword
- can't be `true`, `false` or `null`
- can be of any length
- remember Java is case sensitive

### Method & Class Abstraction

#### Method Abstraction

Method Abstraction
: When you manage to separate the use of a method from its implementation

> The client can use a method without knowing how it is implemented
>
> > This is known as **_Information Hiding_** or **_Encapsulation_**

```{tip}
You can think of this as the {bdg-white-outline}`Method Header` which is **visible** to the client program but the {bdg-black-outline}`Method Body` is hidden from the Client

{bdg-primary}`In other words the method body is a black box`
```

#### Class Abstraction

Class Abstraction
: When you manage to separation of its implementation from the use of the class

> User of the class does not need to know implementation details of the class for this reason a class is known as a **abstract data type**

### Stepwise Refinement

Also known as **Divide and Conquer** this means that as we are coding we are sub dividing the problem into smaller and smaller more manageable problems

```{tip}
You can use incomplete method called **stubs** in order to code and fill in later
```

#### Benefits

- **Simpler Programs** - smaller methods means program is much more simpler than for example procedure
- **Reusing Methods** - promotes code reuse methods can be invoked multiple times
- **Easier Developing, Debugging and Testing** - each subproblem or method can be developed, tested and then debugged individually. Errors can be isolated
- **Facilitation of Teamwork** - sub-problems or methods can be assigned to different programmers

### Selection Sort

> Here we use the intuitive process of swapping elements

1. Find the first ordered element and move it to the front of the Array

## Abstract Classes & Interfaces

Both Abstract Classes and Interfaces cannot be instantiated with the new keyword

### Abstract Methods

A abstract method is a method <u>without</u> and implementation\
You can think of it as a incomplete class

> Abstract methods are non-static and come without a method's body

### Abstract Classes

> An **Abstract class** cannot be used to create objects. An abstract class can contain abstract methods that are implemented in concrete subclasses

**In the Inheritance Hierarchy**\
Classes become more and more _specific_ and _concrete_ with each new subclass\
Conversely\
Classes become less and less _specific_ and more _general_ as we move up the superclasses

--> When a superclass becomes so abstract it cannot be used to create any specific instances such a class is referred to as an **abstract class**

#### Feature List about Abstract Class

- ##### Abstract Methods in Abstract Class
  **<u>An abstract method cannot be contained in a non-abstract class</u>**\
  If a subclass of an abstract superclass does not implement all the abstract methods, then the subclass must be defined as abstract **All abstract methods must be implemented by the subclass**
- ##### Objects cannot be created from Abstract Class
  Cannot be instantiated with a `new` operator, _but you can still define its constructors_ for the subclasses to be invoked by the subclasses
- ##### Abstract Class without abstract methods
  **A class that contains abstract methods must be abstract**\
  However, it is possible to define an abstract class that doesnâ€™t contain any abstract methods\
  This abstract class is used as a base class for defining subclasses
- ##### Concrete Method Overridden to be abstract
  **A subclass can override a method from its superclass to define it as abstract**\
- ##### Subclass Abstract but Superclass Concrete
  **A subclass can be abstract even if its supper-class is concrete**\
  eg `Object` is concrete, but its subclasses such as `GeometricObject` may be abstract
- ##### Abstract Class as a Type
  **You cannot create an instance from an abstract class using the `new` operator, but an abstract class can be used as a data type**\
  Therefore following statement is correct

```java
GeometricObject[] objects = new GeometricObject[10];
// you can then create a instance of GeometricObject and assign its reference to the array like this
object[0] = new Circle();
```

##### Geometric Object

`GeometricObject` models common features of geometric object\
Both `Rectangle` and `Circle` contain the `getArea()` and `getPerimeter()` methods\

Since you can compute areas and parameters of all geometric areas\
it is better to define these methods in the `GeometricObject` class

These methods cannot be implemented in a `GeometricObject` class because their implementation depends on the specific type of geometric object

Such methods are referred to as **abstract methods** and are denoted using the `abstract` modifier in the method header

Once you define the abstract methods in `GeometricObject` it becomes a **abstract class**\
These are denoted using the `abstract` modifier in the class header


`````java
public abstract class GeometricObject {
  private String color = "white";
  private boolean filled;
  private java.util.Date dateCreated;

  /** Construct a default geometric object */
  protected GeometricObject() {
    dateCreated = new java.util.Date();
  }

  /** Construct a geometric object with color and filled value */
  protected GeometricObject(String color, boolean filled) {
    dateCreated = new java.util.Date();
    this.color = color;
    this.filled = filled;
  }

  public String getColor() {
    return color;
  }
  public void setColor(String color) {
    this.color = color;
  }

  /** Return filled. Since filled is boolean,
   *  the get method is named isFilled */
  public boolean isFilled() {
    return filled;
  }

  /** Set a new filled */
  public void setFilled(boolean filled) {
    this.filled = filled;
  }
  /** Get dateCreated */
  public java.util.Date getDateCreated() {
    return dateCreated;
  }

  @Override
  public String toString() {
    return "created on " + dateCreated + "\ncolor: " + color +
      " and filled: " + filled;
  }

  /** Abstract method getArea */
  public abstract double getArea();

  /** Abstract method getPerimeter */
  public abstract double getPerimeter();
}

public class Circle extends GeometricObject {
  private double radius;

  public Circle() {
  }

  public Circle(double radius) {
    this.radius = radius;
  }

  public double getRadius() {
    return radius;
  }
  public void setRadius(double radius) {
    this.radius = radius;
  }
  public double getDiameter() {
    return 2 * radius;
  }

  // Here is how you implement the abstract methods
  @Override /** Return area */
  public double getArea() {
    return radius * radius * Math.PI;
  }
  @Override /** Return perimeter */
  public double getPerimeter() {
    return 2 * radius * Math.PI;
  }

  public void printCircle() {
    System.out.println("The circle is created " + getDateCreated() +
      " and the radius is " + radius);
  }

  @Override /** Return a string representation of a Circle object */
  public String toString() {
    return super.toString() + " radius: " + radius;
  }
}

public class Rectangle extends GeometricObject {
  private double width;
  private double height;

  public Rectangle() {
  }

  public Rectangle(double width, double height) {
    this.width = width;
    this.height = height;
  }

  public double getWidth() {
    return width;
  }
  public void setWidth(double width) {
    this.width = width;
  }

  public double getHeight() {
    return height;
  }
  public void setHeight(double height) {
    this.height = height;
  }

  // Here is how you implement the abstract methods
  @Override /** Return area */
  public double getArea() {
    return width * height;
  }
  @Override /** Return perimeter */
  public double getPerimeter() {
    return 2 * (width + height);
  }

  @Override /** Return a string representation of a Rectangle object */
  public String toString() {
    return super.toString() + " width: " + width + " height: " + height;
  }
}

public class TestGeometricObject {
  public static void main(String[] args) {
    GeometricObject geoObject1 = new Circle(5);
    GeometricObject geoObject2 = new Rectangle(5, 3);

    System.out.println("The two objects have the same area? " +
      equalArea(geoObject1, geoObject2));

    displayGeometricObject(geoObject1);
    displayGeometricObject(geoObject2);
  }

  /** A method for comparing the areas of two geometric objects */
  public static boolean equalArea(GeometricObject object1,
      GeometricObject object2) {
    return object1.getArea() == object2.getArea();
  }

  /** A method for displaying a geometric object */
  public static void displayGeometricObject(GeometricObject object) {
    System.out.println();
    System.out.println("The area is " + object.getArea());
    System.out.println("The perimeter is " + object.getPerimeter());
  }
}
TestGeometricObject.main(new String[0]);
`````

### Inner-Classes

These are defined within a outer-class

An inner class can reference the data and methods
defined in the outer class in which it nests, so you do
not need to pass the reference of the outer class to
the constructor of the inner class

- always uses the **no-arg contractor** from its superclass to create an instance
- if an anonymous inner class implements an interface the constructor is `Object()`
- always compiled into class named `OuterClassName$n.class`
  - example to anonymous classes of outer class `Test` compiled into `Test$1.class` and `Test$2.class`

An inner class is compiled into a class _named_
`OuterClassName$InnerClassName.class`

**Example** the inner class `B` in the outer class
`A` is compiled into `A$B.class`

#### Anonymous Inner-Classes

These are inner-classes **with no names**

They must <u>always</u> extend a superclass or implement an interface but it cannot have a explicit extends or implement clause so they must implement all abstract methods in the superclass or in the interface

In the example the portion

```java
new HelloWorld() {
    @Override
    public void greet() {
        System.out.println("Hello World!");
    }
};
```

Is the unnamed anonymous class even though it is referenced by the `helloWorld` variable


`````java
public interface HelloWorld {
  public void greet();
}

public class MainClass {
  public static void main(String[] args) {
      // Anonymous inner-class implementation of HelloWorld interface
      HelloWorld helloWorld = new HelloWorld() { 
          @Override
          public void greet() {
              System.out.println("Hello World!");
          }
      };
      
      helloWorld.greet();  // Outputs "Hello World!"
  }
}

MainClass.main(new String[0]);
`````

### Listener for Observable Objects

You can add a listener to process a value change in an observable object

## Binary IO

We have already learned about textIO it is time to learn about <strong><u>BinaryIO</u></strong>

- Data stored in a text file is represented in human-readable
  form
- Data stored in a binary file is represented in binary
  form-
- You cannot read binary files -
- They are designed to be read by programs
  - For example, Java source programs are stored in text files and can be read by a text editor, but
- Java classes are stored in **binary files** and are read by the
  JVM

> The advantage of binary files is that they are <u>more efficient</u> to process than text files

```Java
// Input Object created from an input class
// <-- Input Stream (0101001...101001) <-- File
Scanner input = new Scanner(new File("temp.txt"));
System.out.println(input.nextLine());

// Output Object created from an output class
// --> Output Stream (0101001...101001) --> File
PrintWriter output = new PrintWriter("temp.txt");
output.println("Java 101");
output.close();
```

_Text I/O_ requires **encoding** and **decoding**
The JVM converts a Unicode to a file specific encoding when writing a character and coverts a file specific encoding to a Unicode when reading a
character.

<big>Binary I/O does not require conversions</big><br>
When you write a byte to a file, the original byte is copied into the file <br>
When you read a byte from a file, the exact byte in the file is returned <br>

### Filter Input and Filter Output Stream

and characters

`FilterInputStream` and `FilterOutputStream` are the base classes for
filtering data.

**When you need to process primitive numeric types**, use `DataInputStream`
and `DataOutputStream` to filter bytes

### Comparing Reference and Array-Based ADT

A reference-based linked list, also known as a singly-linked list, and an array-based linked list are two different implementations of the linked list data structure.

#### Advantages of a Reference-Based Linked List:

- **Dynamic Size:** A reference-based linked list can grow or shrink dynamically without needing to predefine its size like an array-based implementation. This makes it more flexible for handling varying amounts of data
- **Efficient Insertion and Deletion:** Inserting or deleting elements at any position in a reference-based linked list is efficient, typically requiring only updating the references (pointers) between nodes. It doesn't involve shifting elements like in an array
- **Memory Efficiency:** A reference-based linked list uses memory more efficiently than an array since it only allocates memory for the actual elements stored and their associated pointers, rather than allocating memory for a fixed-size block regardless of how many elements are present

#### Disadvantages of a Reference-Based Linked List:

- **Sequential Access Only:** Unlike arrays that allow random access based on indices, accessing elements in a reference-based linked list requires traversing through the links from one node to another sequentially until reaching the desired element. This can make searching or accessing specific items less efficient compared to arrays
- **Extra Memory Overhead:** Each node in a reference-based linked list requires extra memory to store both the data element and the pointer/reference to the next node in line. This additional overhead can increase memory consumption compared to arrays when dealing with large datasets
- **Lack of Random Access:** Since each element is accessed sequentially starting from the head/first node, there is no direct way to access arbitrary positions within the referenced based linked lists without traversing through all preceding nodes first

#### Advantages of Array-Based Linked List:

- **Random Access Capability:** Elements can be accessed directly using indices because they are stored contiguously within an underlying array-like structure similar to regular arrays
- **Memory Efficiency:** Array-based linked lists can be more memory-efficient than reference-based linked lists, especially for small and fixed-size collections of elements since they don't require extra pointers between nodes

#### Disadvantages of Array-Based Linked List:

- **Fixed Size:** The size of an array-based linked list is fixed upon creation, requiring resizing operations if the number of elements exceeds its initial capacity. This makes it less flexible when handling dynamic data or frequently changing collection sizes
- **Costly Insertions and Deletions:** Inserting or deleting elements within an array-based linked list requires shifting the remaining elements to accommodate the changes, which can be computationally expensive for large arrays
- **Inefficient Memory Usage:** If the array's capacity is not fully utilized, there may be wasted memory space since it reserves a fixed block regardless of actual element count

In summary, a reference-based linked list offers **dynamic sizing** and **efficient insertions/deletions** at the cost of **sequential access** and **additional memory overhead per node**. On the other hand, an array-based implementation provides **random access capability** but has a **fixed size** and _potentially inefficient memory usage due to reserved block allocation_

`````java
import java.util.Arrays;

class BubbleSort {
  
  static int[] bubbleSort(int[] arr) {
    
    boolean isSorted = true;
    for (int i = 0; i < arr.length - 1; i++) {
      if (arr[i] > arr[i + 1]) {
        // swap values
        int num1 = arr[i];
        int num2 = arr[i + 1];
        arr[i] = num2;
        arr[i + 1] = num1;
        isSorted = false;
      }
    }
    // if pass through array without swapping then sort complete else recursively
    // sort array again in line with the bubble sort algorithm
    
    System.out.println("BubbleSort:\tafter pass => " + 
    ":\t" + Arrays.toString(arr) );
    
    return isSorted ? arr : bubbleSort(arr);
  }
}

int[] arr = {29, 10, 14, 37, 13};
BubbleSort.bubbleSort(arr);
`````

#### Heap Sort

Heap sort is a comparison-based sorting algorithm that uses a binary heap data structure to sort elements efficiently. Here's an explanation of the steps involved in the heap sort algorithm:

**Build Max-Heap**: The first step is to build a max-heap from the given array. This involves rearranging the elements of the array so that they satisfy the max-heap property, which states that for any node i, its value should be greater than or equal to its child nodes' values.

**Extract Maximum and Heapify:** After building the max-heap, we repeatedly extract the maximum element from it and restore the max-heap property. To do this, we swap the root (maximum element) with the last leaf node of the heap, reduce its size by one, and then call a procedure called "Heapify" on the root node to maintain its property.

**Repeat Extraction:** We repeat step 2 until all elements have been extracted from the heap. As each maximum element is extracted, it gets placed at t

`````java
/* TIGHT VS LOOSE COUPLING
 * note instead of getters and setters we us public field declaration to save space
 */

// TIGHTLY COUPLED

public class Item {
  public String name;
  public float price;
  public int quantity;
}

public class Cart {
  public Item[] items;
}

public class Checkout {
  private Cart cart; // checkout needs to know that cart is a array of Items
  private float discount;
  public Checkout(Cart cart, float disc) {
    this.cart = cart;
    discount = disc;
  }
  public float totalPrice() {
    float total = 0;
    for(int i = 0; i < cart.items.length; ++i)
      total += cart.items[i].price * cart.items[i].quantity;     
    total = total * (1 - discount);
    return total;
  }
}

Item item1 = new Item();
item1.name = "Item1";
item1.price = (float) 50.01;
item1.quantity = 2;

Item item2 = new Item();
item2.name = "Item2";
item2.price = (float) 10.10;
item2.quantity = 10;

Cart cart = new Cart();
cart.items = new Item[]{item1, item2};

Checkout checkout = new Checkout(cart, (float) 0.10);
System.out.print("\nCheckout total price:\t" + checkout.totalPrice());
System.out.print("\t (Loosely Coupled)");

// moreover checkout needs to know that Item have a price and a quantity

// LOOSELY COUPLED
public class Item {
  public String name;
  public float price;
  public int quantity;

  public float getTotalCost() {
    return price * quantity;
  }
}

public class Cart {
  public Item[] items;

  public float getCartValue() {
    float total = 0;
    for(int i = 0; i < items.length; ++i)
      total += cart.items[i].getTotalCost();     
    return total;
  }
}

public class Checkout {
  private Cart cart; // checkout doesn't need to know that cart is a array of Items
  private float discount;

  public Checkout(Cart cart, float disc) {
    this.cart = cart;
    discount = disc;
  }
  public float toPay() {
    return cart.getCartValue() * (1 - discount);
  }
}

Item item1 = new Item();
item1.name = "Item1";
item1.price = (float) 50.01;
item1.quantity = 2;

Item item2 = new Item();
item2.name = "Item2";
item2.price = (float) 10.10;
item2.quantity = 10;

Cart cart = new Cart();
cart.items = new Item[]{item1, item2};

Checkout checkout = new Checkout(cart, (float) 0.10);
System.out.print("\nCheckout total price:\t" + checkout.toPay());
System.out.print("\t (Tightly Coupled)");

// now checkout doesn't need to know that Item have a price and a quantity because of methods
`````
